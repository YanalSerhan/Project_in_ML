```python
import json
import csv

# Load JSON file
with open("Tcomments.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Extract table rows (the `data` field inside the table entry)
table_entry = next(item for item in data if item.get("type") == "table")
rows = table_entry["data"]

# Get CSV headers from the keys of the first row
headers = rows[0].keys()

# Write CSV file
with open("comments.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader()
    writer.writerows(rows)

print("CSV file created as comments.csv")
```
